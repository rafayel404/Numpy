```python

# plotting 3D graphs
# meshgrids
a = np.linspace(-10,9,20)
b = np.linspace(-10,9,20)
b
```

```python
xx,yy=np.meshgrid(a,b)
```

```python
import matplotlib.pyplot as plt
plt.scatter(xx,yy)
```

```python
def func(x,y):
  return xx**2+yy*2

zz = func(xx,yy)
zz
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = px.scatter_3d()

fig.add_trace(go.Surface(x=xx,y=yy,z=zz))

fig.show()
```

```python
# name,iq,cgpa,placed

dt = np.dtype(
    [
        ('name','U20'),
        ('iq',np.int32),
        ('cgpa',np.float64),
        ('placed','U20')
    ]
)
```
